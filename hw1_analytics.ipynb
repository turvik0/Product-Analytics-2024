**Для структурирования метрик будем использовать следующие фреймворки:**


1.   AARRR (Pirate Metrics):
    * Acquisition (Привлечение)
    * Activation (Активация)
    * Retention (Удержание)
    * Referral (Рекомендации)
    * Revenue (Доход)

2.   HEART Framework:
  * Happiness (Удовлетворенность)
  * Engagement (Вовлеченность)
  * Adoption (Принятие)
  * Retention (Удержание)
  * Task Success (Успех выполнения задач)

**Основные метрики для дашборда:**

* DAU (Daily Active Users): Количество уникальных пользователей, которые используют приложение ежедневно.Нагша сновная метрика, которую мы хотим растить. Показывает текущую активность пользователей.

* MAU (Monthly Active Users): Количество уникальных пользователей, которые используют приложение ежемесячно. Помогает понять долгосрочную активность пользователей.

* Stickiness Ratio (DAU/MAU): Отношение DAU к MAU. Показывает, насколько часто пользователи возвращаются в приложение.

* New User Acquisition: Количество новых пользователей, которые установили приложение. Помогает отслеживать эффективность маркетинговых кампаний и стратегий привлечения.

* Retention Rate: Процент пользователей, которые возвращаются в приложение через определенный период времени.Показывает как хорошо получается удерживать пользователей.

* Engagement Metrics:
  * Average Session Length: Средняя продолжительность сессии.
  * Messages Sent: Количество отправленных сообщений.
  * Channels/Groups Joined: Количество каналов и групп, к которым * присоединились пользователи.
Все они показывают уровень вовлеченности пользователей в приложение.

* User Satisfaction (NPS, CSAT): Оценка удовлетворенности пользователей (Net Promoter Score, Customer Satisfaction Score), помогает понять, насколько пользователи довольны приложением и готовы ли они его рекомендовать.

* Task Success Rate: Процент успешно выполненных пользовательских задач (например, отправка сообщения, присоединение к группе).Показывает эффективность и удобство использования приложения.

* Referral Rate:Процент пользователей, которые пригласили новых пользователей.Помогает понять, насколько пользователи готовы рекомендовать приложение другим.

* Revenue Metrics:
  * Total Revenue: Общий доход.
  * ARPU (Average Revenue Per User): Средний доход на пользователя.
  * Почему: Показывают финансовую эффективность приложения.
